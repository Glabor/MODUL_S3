```python
fn="hw"
with open(f"{fn}.bin", 'rb') as file:
    content=file.readlines()
    content = b''.join(content)
```

```python
import pandas as pd
import plotly.express as px

tab_splitted=content.split(b'\n\r\t')[1:]
# print(tab_splitted[0:100])
lt,lx,ly,lz=[],[],[],[]
for t in tab_splitted:
    # print(len(t))
    lt.append(int.from_bytes(t[0:4],byteorder='big', signed=False))  
    lx.append(int.from_bytes(t[4:7],byteorder='big', signed=False))  
    ly.append(int.from_bytes(t[7:10],byteorder='big', signed=False))  
    lz.append(int.from_bytes(t[10:13],byteorder='big', signed=False))  

# for i in range(0,len(content),param[fn]):
#     lt.append(int.from_bytes(content[i:i+4],byteorder='big', signed=False))  
#     if fn=="sick":
#         lx.append(int.from_bytes(content[i+4:i+6],byteorder='big', signed=True))
#     else:
#         lx.append(int.from_bytes(content[i+4:i+6],byteorder='big', signed=True)/100)
#         ly.append(int.from_bytes(content[i+6:i+8],byteorder='big', signed=True)/100)
#         lz.append(int.from_bytes(content[i+8:i+10],byteorder='big', signed=True)/100)
    # print(t,x/100,y/100,z/100)
```

```python
lt2,lx2,ly2,lz2=[],[],[],[]
for i in range(0,len(lt)-2, 2):
    lt2.append((lt[i]+lt[i+1])/2)
    lx2.append((lx[i]+lx[i+1])/2)
    ly2.append((ly[i]+ly[i+1])/2)
    lz2.append((lz[i]+lz[i+1])/2)
```

```python
df=pd.DataFrame((lt2,lx2,ly2,lz2)).transpose()
df.columns=("t","x","y","z")
df_a=df[(df["t"]>0.0)&(df["t"]<300.0e6)]
```

```python
fig =px.line(None, x=[0], y=[0], title="hw measure", markers=True)
fig =fig.add_scatter(x=df_a["t"], y=df_a["x"], mode="lines+markers")
# if fn!="sick":
fig =fig.add_scatter(x=df_a["t"], y=df_a["y"], mode="lines+markers")
fig =fig.add_scatter(x=df_a["t"], y=df_a["z"], mode="lines+markers")
fig.update_layout( showlegend=True)
fig.show()
```
